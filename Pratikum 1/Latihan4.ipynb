```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.datasets import fetch_california_housing
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


housing = fetch_california_housing()
df = pd.DataFrame(housing.data, columns=housing.feature_names)
df['MedHouseValue'] = pd.Series(housing.target)

target = df['MedHouseValue']
features = df.drop('MedHouseValue', axis=1)
df.tail()

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.1, random_state=0)

scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
print(X_test)

model = LinearRegression()
model.fit(X_train, y_train)

print("Intercept: ", model.intercept_)
print("Slope (Coefficients): ", model.coef_)

rmse_train = np.sqrt(mean_squared_error(y_train, model.predict(X_train)))
rmse_test = np.sqrt(mean_squared_error(y_test, model.predict(X_test)))

r2_train = model.score(X_train, y_train)
r2_test = model.score(X_test, y_test)

print("RMSE Train: ", rmse_train)
print("RMSE Test: ", rmse_test)
print("R^2 Train:", r2_train)
print("R^2 Test: ", r2_test)
```
